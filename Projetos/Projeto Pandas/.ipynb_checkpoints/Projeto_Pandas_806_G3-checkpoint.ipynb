````python
#Pandas
from IPython.display import Image, display
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from pandas.api.types import is_string_dtype
from pandas.api.types import is_numeric_dtype
````

````python
#Questão 1
#Importar todos os dados
df3 = pd.read_csv('3_bens.csv',sep=";")
df2 = pd.read_csv('2_renda_gastos.csv',sep=";")
df1 = pd.read_csv('1_demografico.csv',sep=";")

#Concatenar todos os dados formando um só DataFrame
df = pd.concat([df1, df2, df3], axis=1)

#Eliminar todas as colunas iguais
#df = df.T.drop_duplicates().T
df = df.loc[:,~df.columns.duplicated()]
df
````

````python
#Questão 2 - 1
#Lista de cabeças do dataframe
heads = list(df.columns.values)
print(heads)
````

````python
#Questão 2 - 2
#Faca um relatório no qual cada variável quantitativa deve ter:
#Mínimo | Primeiro Quartil | Mediana |Terceiro Quartil | Máximo | Média

#Listas vazias para alocar os valores
nomes, media, maximo, minimo, mediana, q1, q3 = [],[],[],[],[],[],[]
 
def criar_dados(nome):
    #Funciona somente se os dados da coluna não forem do tipo Objeto (Equivalente a String)
    if df[nome].dtype != 'O':
        #Criação de todos os dados solicitados
        nomes.append(nome)
        maximo.append(df[nome].max())
        minimo.append(df[nome].min())
        q1.append(df[nome].quantile(q=0.25))
        q3.append(df[nome].quantile(q=0.75))
        mediana.append(df[nome].median())
        media.append(round(df[nome].mean(),2))

#Para cada coluna será executada a função
for nome in heads[1:]:
    criar_dados(nome)

dados = {
    'ID': nomes,
    'Média': media,
    'Máximo': maximo,
    'Minímo': minimo,
    'Q1': q1,
    'Mediana': mediana,
    'Q3': q3,
}

rel = pd.DataFrame(dados)
rel
````

````python
#Questão 3 - 1
#Dataframe 3 (D3) vai ser uma cópia do Df original
d3 = df
````

````python
#Questão 3 - 2
#Enunciado:Trate os dados Outliers como faltantes e crie novamente o relatório para observarmos a diferenca

nomes, media, maximo, minimo, mediana, q1, q3 = [],[],[],[],[],[],[]

def remover_outlier(nome):
    #Transforma em np.nan todos os Outliers 
    if d3[nome].dtype != 'O':
        #Cálculo dos limites superior e inferior
        mediana = d3[nome].median()
        iqr = d3[nome].quantile(q=0.75)-d3[nome].quantile(q=0.25)
        superior = mediana + 1.5 * iqr
        inferior = mediana - 1.5 * iqr

        for i in range(d3.shape[0]):
            #Aplicar para cada item de cada coluna a lógica
            if d3.loc[i,nome]>superior or d3.loc[i,nome]<inferior:
                d3.loc[i,nome] = np.nan
        
def criar_dados(nome):
    if d3[nome].dtype != 'O':
        nomes.append(nome)
        maximo.append(d3[nome].max())
        minimo.append(d3[nome].min())
        q1.append(d3[nome].quantile(q=0.25))
        q3.append(d3[nome].quantile(q=0.75))
        mediana.append(d3[nome].median())
        media.append(round(d3[nome].mean(),2))
        
dados = {
    'ID': nomes,
    'Média': media,
    'Máximo': maximo,
    'Minímo': minimo,
    'Q1': q1,
    'Mediana': mediana,
    'Q3': q3,
}
       
for nome in heads[1:]:
    remover_outlier(nome)

for nome in heads[1:]: 
    criar_dados(nome)    
````

````python
#Questão 3 - 3
#Formar o data frame com os dados e o exibir
dataframe3 = pd.DataFrame(dados)
dataframe3
````

````python
#Questão 4 - 1
#Dataframe 4 (D4) vai ser uma cópia do Df original
d4 = df
````

````python
#Questão 4 - 2
#4)Trate os dados Outliers inserindo no lugar a mediana e crie novamente o relatório para observarmos a diferenca:

nomes, media, maximo, minimo, mediana, q1, q3 = [],[],[],[],[],[],[]

def outlier_para_mediana(nome):
    #Transforma todos os outliers em mediana 
    if d4[nome].dtype != 'O':
        #Cálculo dos limites superior e inferior
        mediana = d4[nome].median()
        iqr = d4[nome].quantile(q=0.75)-d4[nome].quantile(q=0.25)
        superior = mediana + 1.5 * iqr
        inferior = mediana - 1.5 * iqr

        for i in range(d3.shape[0]):
            #Aplicar para cada item de cada coluna a lógica
            if d4.loc[i,nome]>superior or d4.loc[i,nome]<inferior:
                d4.loc[i,nome] = mediana
        
def criar_dados(nome):
    if d4[nome].dtype != 'O':
        nomes.append(nome)
        maximo.append(d4[nome].max())
        minimo.append(d4[nome].min())
        q1.append(d4[nome].quantile(q=0.25))
        q3.append(d4[nome].quantile(q=0.75))
        mediana.append(d4[nome].median())
        media.append(round(d4[nome].mean(),2))
        
dados = {
    'ID': nomes,
    'Média': media,
    'Máximo': maximo,
    'Minímo': minimo,
    'Q1': q1,
    'Mediana': mediana,
    'Q3': q3,
}
       
for nome in heads[1:]:
    outlier_para_mediana(nome)

for nome in heads[1:]: 
    criar_dados(nome)    
````

````python
#Questão 4 - 3
#Formar o data frame com os dados e o exibir
dataframe4 = pd.DataFrame(dados)
dataframe4
````

````python
#Questão 5 - 1
#Exibe quais são as heads para atributos qualitativos
head_objetos = []
for nome in heads:
    if df[nome].dtype == 'O':
        head_objetos.append(nome)
print(head_objetos)       
````

````python
#Questão 5 - 2
#Inserir o nome da head qualitativa

for (index, coluna) in enumerate(df):
    if df[coluna].dtype =='object':
        dfAbs = df[coluna].value_counts()
        dfRel = df[coluna].value_counts()/len(df)
        dfAbsAcc = dfAbs.cumsum(axis = 0)
        dfRelAcc = dfRel.cumsum(axis = 0)
        dfRelPerc = dfRel * 100
        dfFreqs = pd.concat([dfAbs, dfAbsAcc, dfRel, dfRelPerc, dfRelAcc], axis=1)
        dfFreqs = dfFreqs.set_axis(['Abs.', 'Abs. Acum.', 'Rel.', 'Rel. %', 'Rel. Acum.'], axis=1, inplace=False)
        
        print('\n\n', coluna)
        print(dfFreqs)
````

````python
# Questão 6
maxRecord = int(len(df)*0.1)
df10 = df.sort_values(['Total Household Income'], ascending = False).reset_index().iloc[:maxRecord]
df10.to_csv('maioresRendas.csv', index = False)
````

````python
# Questão 7
mainSourceOfIncome_dummies = pd.get_dummies(df['Main Source of Income'], prefix = 'MainSourceOfIncome' )
mainSourceOfIncome_dummies.sample(n = 5, random_state = 1 )
df = pd.concat([df,mainSourceOfIncome_dummies], axis = 1)

householdHeadSex_dummies = pd.get_dummies(df['Household Head Sex'], prefix = 'HouseholdHeadSex' )
householdHeadSex_dummies.sample(n = 5, random_state = 1 )
df = pd.concat([df, householdHeadSex_dummies], axis = 1)

householdHeadMaritalStatus_dummies = pd.get_dummies(df['Household Head Marital Status'], prefix = 'HouseholdHeadMaritalStatus' )
householdHeadMaritalStatus_dummies.sample(n = 5, random_state = 1 )
df = pd.concat([df, householdHeadMaritalStatus_dummies], axis = 1)

# Para visualizar 5 linhas quaister
df.sample(n=5, random_state=1)
````

````python
# Recarrega o dataset das 10% maiores rendas com as variáveis dummies
df10 = df.sort_values(['Total Household Income'], ascending = False).reset_index().iloc[:maxRecord]
````

````python
# Questão 8
def RelatorioMedias(df, df10, coluna):
    print('\n')
    print(coluna)
    print('População em geral:', round(df[coluna].mean(), 2), '\t\t', '10% dos mais ricos:', round(df10[coluna].mean(), 2))
    
    
RelatorioMedias(df, df10, 'Total Number of Family members')
RelatorioMedias(df, df10, 'House Age')
RelatorioMedias(df, df10, 'Number of Cellular phone')
RelatorioMedias(df, df10, 'HouseholdHeadMaritalStatus_Married')
RelatorioMedias(df, df10, 'HouseholdHeadSex_Female')
RelatorioMedias(df, df10, 'MainSourceOfIncome_Enterpreneurial Activities')
````
